```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('TWO_CENTURIES_OF_UM_RACES.csv')
```

```python
df.head(5)
```

```python
df.shape
```

```python
df.dtypes
```

### Clean up data

```python
df[(df['Event distance/length']=='50km')|(df['Event distance/length']=='50mi')|(df['Event distance/length']=='100km')|(df['Event distance/length']=='100mi')]
```

```python
df[(df['Event distance/length'].isin(['50km','50mi','100km','100mi'])) & (df['Year of event']==2020)]
```

#### Choosing Country, Year and Distance

```python
df[(df['Event name'].str.contains('(AUS)')) & (df['Year of event']==2022)].sort_values(by='Year of event',ascending=False)
```

```python
df2 = df[df['Event name'].str.contains('(AUS)') & (df['Year of event']==2022)]
df2.head(5)
```

```python
df2 = df2[(df2['Event distance/length']=="50mi") | (df2['Event distance/length']=="50km") | (df2['Event distance/length']=="100km") | (df2['Event distance/length']=="100mi")]
df2.head(5)
```

#### Fix column "Event name"

```python
df2["Event name"] = df2["Event name"].str.split('(').str.get(0)
df2.head(5)
```

#### Fix athlete age

```python
df2["Athlete age"] = df2["Year of event"] - df2["Athlete year of birth"]
```

#### Fix column 'Athlete performance'

```python
df2["Athlete performance"] = df2["Athlete performance"].str.split(' ').str.get(0)
```

#### Drop columns

```python
df2 = df2.drop(['Athlete club','Athlete year of birth','Athlete age category'],axis=1)
df2.head(5)
```

#### Clean up null values

```python
df2.isna().sum()
```

```python
df2[df2["Athlete age"].isna()==1]
```

```python
df2=df2.dropna()
```

```python
df2.shape
```

#### Duplicate values

```python
df2[df2.duplicated()==True]
```

#### Fix index

```python
df2.reset_index(drop=True)
```

#### Fix Dtype

```python
df2.dtypes
```

```python
df2["Athlete age"]=df2["Athlete age"].astype(int)
```

```python
df2["Athlete average speed"]=df2["Athlete average speed"].astype(float)
```

#### Remane columns

```python
df2=df2.rename(columns={
    'Year of event'            :'year',                  
    'Event dates'              :'date', 
    'Event name'               :'race_name', 
    'Event distance/length'    :'race_length', 
    'Event number of finishers':'race_number_finishers', 
    'Athlete performance'      :'athlete_performance', 
    'Athlete country'          :'athlete_country', 
    'Athlete gender'           :'athlete_gender', 
    'Athlete average speed'    :'athlete_average_speed', 
    'Athlete ID'               :'athlete_id', 
    'Athlete age'              :'athlete_age' 
})
```

```python
df2.head(4)
```

#### Reorder columns

```python
df3 = df2[['date','race_name','race_length','race_number_finishers','athlete_id','athlete_gender','athlete_age','athlete_country','athlete_performance','athlete_average_speed']]
```

```python

df3['date'] = pd.to_datetime(df3['date'], format='%d.%m.%Y', errors='coerce')
```

```python
df3 = df3.dropna()
```

```python
df3.head(4)
```

#### Graphs

```python
sns.histplot(df3['race_length'],binwidth=1)
plt.xlabel('race_length')
plt.ylabel('Count')
plt.title('Race Length')
plt.show()
```

```python
sns.histplot(df3, x='race_length', hue='athlete_gender')
```

```python
sns.displot(df3[df3['race_length']=='50km']['athlete_average_speed']) 
```

```python
sns.histplot(df3['athlete_age'], bins=10)
plt.xlabel('Athleta age')
plt.ylabel('Count')
plt.title('Histograma de Edades de Atletas')
plt.show() 
```

```python
sns.violinplot(x='athlete_gender', y='athlete_average_speed', data=df3, hue= 'athlete_gender', split=True, inner='quart')
plt.xlabel('Atleta_gender')
plt.ylabel('AVG speed')
plt.title('Diagrama de Violín de Velocidad Promedio por Género')
plt.show()
```

```python
sns.lmplot(data=df3,x='athlete_age', y='athlete_average_speed', hue='athlete_gender')
```

#### Questions

```python
#Difference in speed for the 50km, 50mi male to female
df3.groupby(['race_length','athlete_gender'])['athlete_average_speed'].mean()
```

```python
# What age group are the best in the race (Best time)
df3.groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=False).query('count>10')
```

```python
# What age group are the best in the race (Worse time)
df3.groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=True).query('count>10')
```

```python
# What nacionality group are the best in the race (Exclude AUS)
df3.groupby('athlete_country')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=True).query('count>10').query('athlete_country!="AUS"')
```

```python
#Count race by gender (Male)
df3.groupby(['race_name','athlete_gender'])['athlete_gender'].agg(['count']).query('athlete_gender == "M"').query('count>1').sort_values('count',ascending=False)
```

```python
#Count race by gender (FEMALE)
df3.groupby(['race_name','athlete_gender'])['athlete_gender'].agg(['count']).query('athlete_gender == "F"').query('count>1').sort_values('count',ascending=False)
```

```python
#Count and avg in race length by gender 
df3.groupby(['race_length','athlete_gender'])['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=False).query('count>10')
```
